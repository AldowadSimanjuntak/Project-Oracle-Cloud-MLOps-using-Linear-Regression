```scala
%python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
data = df

# Mengurutkan negara berdasarkan jumlah populasi pada tahun 2020
data_sorted = data.sort_values(by='YEAR_2020', ascending=False)

# Mengambil 10 negara dengan jumlah terbanyak
top_10_countries = data_sorted.head(10)

# Mengambil kolom-kolom tahun yang diinginkan
years = ['YEAR_1960', 'YEAR_1970', 'YEAR_1980', 'YEAR_1990', 'YEAR_2010', 'YEAR_2020']

# Mengambil data populasi untuk 10 negara tersebut pada tahun-tahun yang diinginkan
populations = top_10_countries[years].values

# Visualisasi grafik
plt.figure(figsize=(20, 8))
for i in range(len(top_10_countries)):
    plt.plot(years, populations[i], label=top_10_countries.iloc[i, 0])

plt.xlabel('Tahun')
plt.ylabel('Jumlah Populasi')
plt.title('Top 10 Negara dengan Jumlah Populasi Terbanyak')
plt.legend()
plt.show()

```

```scala
%python
# Mengambil kolom negara dan kolom tahun-tahun yang diinginkan
selected_columns = ['COUNTRY'] + [f'YEAR_{year}' for year in range(1960, 2021)]
selected_data = data[selected_columns]
```

```scala
%python
# Memilih data untuk negara yang dipilih
countries = ['China', 'India', 'United States', 'Indonesia']
selected_data = selected_data[selected_data['COUNTRY'].isin(countries)]
```

```scala
%python
print(selected_data)
```

```scala
%python
# Memisahkan data menjadi fitur (X) dan target (y)
X = selected_data.iloc[:, 1:].values  # Mengambil semua kolom kecuali kolom pertama (negara)
y = selected_data['YEAR_2020'].values
```

```scala
%python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
# Membangun model Regresi Linear
model = LinearRegression()
model.fit(X, y)
```

```scala
%python
# Evaluasi model
r_squared = model.score(X, y)
print('R-squared:', r_squared)
```

```scala
%python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Memprediksi target menggunakan fitur X
y_pred = model.predict(X)

# Menghitung Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print('Mean Squared Error (MSE):', mse)

# Menghitung Mean Absolute Error (MAE)
mae = mean_absolute_error(y, y_pred)
print('Mean Absolute Error (MAE):', mae)
```
